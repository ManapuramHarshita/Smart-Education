```python
import random

# Dictionary of states and their capitals
states_capitals = {
    "Andhra Pradesh": "Amaravati",
    "Arunachal Pradesh": "Itanagar",
    "Assam": "Dispur",
    "Bihar": "Patna",
    "Chhattisgarh": "Raipur",
    "Goa": "Panaji",
    "Gujarat": "Gandhinagar",
    "Haryana": "Chandigarh",
    "Himachal Pradesh": "Shimla",
    "Jharkhand": "Ranchi",
    "Karnataka": "Bengaluru",
    "Kerala": "Thiruvananthapuram",
    "Madhya Pradesh": "Bhopal",
    "Maharashtra": "Mumbai",
    "Manipur": "Imphal",
    "Meghalaya": "Shillong",
    "Mizoram": "Aizawl",
    "Nagaland": "Kohima",
    "Odisha": "Bhubaneswar",
    "Punjab": "Chandigarh",
    "Rajasthan": "Jaipur",
    "Sikkim": "Gangtok",
    "Tamil Nadu": "Chennai",
    "Telangana": "Hyderabad",
    "Tripura": "Agartala",
    "Uttar Pradesh": "Lucknow",
    "Uttarakhand": "Dehradun",
    "West Bengal": "Kolkata",
}

# Function for Multiple Choice Questions (MCQs)
def mcq_quiz():
    score = 0
    print("MCQ Quiz: Identify the capital of the state.")
    for _ in range(5):  # 5 random questions
        state = random.choice(list(states_capitals.keys()))
        correct_answer = states_capitals[state]
        options = random.sample(list(states_capitals.values()), 3) + [correct_answer]
        random.shuffle(options)
        print(f"\nWhat is the capital of {state}?")
        for idx, option in enumerate(options):
            print(f"{idx + 1}. {option}")
        answer = input("Choose the correct option (1/2/3/4): ")
        if options[int(answer) - 1] == correct_answer:
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {correct_answer}.")
    print(f"Your score in MCQ Quiz is: {score}/5")

# Function for Fill in the Blanks Quiz
def fill_in_the_blanks():
    score = 0
    print("Fill in the Blanks Quiz: Provide the capital of the state.")
    for _ in range(5):  # 5 random questions
        state = random.choice(list(states_capitals.keys()))
        correct_answer = states_capitals[state]
        answer = input(f"\nThe capital of {state} is: ")
        if answer.strip().lower() == correct_answer.strip().lower():
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer is {correct_answer}.")
    print(f"Your score in Fill in the Blanks Quiz is: {score}/5")

# Function for True or False Quiz
def true_or_false():
    score = 0
    print("True or False Quiz: State whether the capital is correct.")
    for _ in range(5):  # 5 random questions
        state, correct_answer = random.choice(list(states_capitals.items()))
        random_answer = random.choice(list(states_capitals.values()))
        if random.choice([True, False]):
            print(f"\nThe capital of {state} is {correct_answer}. (True/False)")
            answer = input("Your answer: ")
            if answer.strip().lower() == "true":
                print("Correct!")
                score += 1
            else:
                print(f"Wrong! The correct answer is True.")
        else:
            print(f"\nThe capital of {state} is {random_answer}. (True/False)")
            answer = input("Your answer: ")
            if answer.strip().lower() == "false":
                print("Correct!")
                score += 1
            else:
                print(f"Wrong! The correct answer is False.")
    print(f"Your score in True or False Quiz is: {score}/5")

# Main function to run the quiz
def run_quiz():
    print("Welcome to the States and Capitals Quiz Competition!")
    print("Choose a quiz format:")
    print("1. Multiple Choice Questions (MCQs)")
    print("2. Fill in the Blanks")
    print("3. True or False")

    choice = input("Enter the number of your choice: ")

    if choice == "1":
        mcq_quiz()
    elif choice == "2":
        fill_in_the_blanks()
    elif choice == "3":
        true_or_false()
    else:
        print("Invalid choice. Please select a valid option.")

# Run the quiz
if __name__ == "__main__":
    run_quiz()
```
